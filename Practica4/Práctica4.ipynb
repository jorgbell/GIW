```python
import sqlite3
import csv

def leerDatosGenerales():
    csvarchivo = open('Tabla1.csv',encoding="iso-8859-1",errors='ignore')
    entrada = csv.reader(csvarchivo, delimiter=";")
    tabla = list(entrada)
    conn=sqlite3.connect("bolsa.sqlite3")
    cur=conn.cursor()
    cur.execute("DROP TABLE IF EXISTS Datos")
    cur.execute("CREATE TABLE Datos (Ticker TEXT UNIQUE, Nombre TEXT, Indice TEXT, Pais TEXT)")
    for fila in tabla[1:]:
        cur.execute("INSERT INTO Datos VALUES (?,?,?,?)", fila)   
    cur.close()
    conn.commit()

def leerDatosIbex():
    csvarchivo = open('Tabla2(IBEX35).csv',encoding="iso-8859-1",errors='ignore')
    entrada = csv.reader(csvarchivo, delimiter=";")
    tabla = list(entrada)
    conn=sqlite3.connect("bolsa.sqlite3")
    cur=conn.cursor()
    cur.execute("DROP TABLE IF EXISTS Ibex")
    cur.execute("CREATE TABLE Ibex (Ticker TEXT REFERENCES Datos(Ticker) UNIQUE, Precio DOUBLE, VarPorcentaje DOUBLE, VarAbsoluto DOUBLE, Max DOUBLE, Min DOUBLE, Volumen INTEGER)")
    for fila in tabla[1:]:
        cur.execute("INSERT INTO Ibex VALUES (?,?,?,?,?,?,?)", fila)   
    cur.close()
    conn.commit()


def leerDatosDAX30():
    csvarchivo = open('Tabla3(DAX 30).csv',encoding="iso-8859-1",errors='ignore')
    entrada = csv.reader(csvarchivo, delimiter=";")
    tabla = list(entrada)
    conn=sqlite3.connect("bolsa.sqlite3")
    cur=conn.cursor()
    cur.execute("DROP TABLE IF EXISTS DAX30")
    cur.execute("CREATE TABLE DAX30 (Ticker TEXT REFERENCES Datos(Ticker) UNIQUE, Precio DOUBLE, VarPorcentaje DOUBLE, VarAbsoluto DOUBLE, Max DOUBLE, Min DOUBLE, Volumen INTEGER)")
    for fila in tabla[1:]:
        cur.execute("INSERT INTO DAX30 VALUES (?,?,?,?,?,?,?)", fila)
    cur.close()
    conn.commit()
    
def leerDatosNasdaq100():
    csvarchivo = open('Tabla4(Nasdaq 100).csv',encoding="iso-8859-1",errors='ignore')
    entrada = csv.reader(csvarchivo, delimiter=";")
    tabla = list(entrada)
    conn=sqlite3.connect("bolsa.sqlite3")
    cur=conn.cursor()
    cur.execute("DROP TABLE IF EXISTS Nasdaq100")
    cur.execute("CREATE TABLE Nasdaq100 (Ticker TEXT REFERENCES Datos(Ticker) UNIQUE, Precio DOUBLE, VarPorcentaje DOUBLE, VarAbsoluto DOUBLE, Max DOUBLE, Min DOUBLE, Volumen INTEGER)")
    for fila in tabla[1:]:
        cur.execute("INSERT INTO Nasdaq100 VALUES (?,?,?,?,?,?,?)", fila)   
    cur.close()
    conn.commit()
    
def leerDatosFtseMib():
    csvarchivo = open('Tabla5(Ftse Mib).csv',encoding="iso-8859-1",errors='ignore')
    entrada = csv.reader(csvarchivo, delimiter=";")
    tabla = list(entrada)
    conn=sqlite3.connect("bolsa.sqlite3")
    cur=conn.cursor()
    cur.execute("DROP TABLE IF EXISTS FtseMib")
    cur.execute("CREATE TABLE FtseMib (Ticker TEXT REFERENCES Datos(Ticker) UNIQUE, Precio DOUBLE, VarPorcentaje DOUBLE, VarAbsoluto DOUBLE, Max DOUBLE, Min DOUBLE, Volumen INTEGER)")
    for fila in tabla[1:]:
        cur.execute("INSERT INTO FtseMib VALUES (?,?,?,?,?,?,?)", fila)   
    cur.close()
    conn.commit()
def leerDatosSemanalesIBEX():
    csvarchivo = open('Tabla6(SemanalesIBEX35).csv',encoding="iso-8859-1",errors='ignore')
    entrada = csv.reader(csvarchivo, delimiter=";")
    tabla = list(entrada)
    conn=sqlite3.connect("bolsa.sqlite3")
    cur=conn.cursor()
    cur.execute("DROP TABLE IF EXISTS SemanalesIBEX35")
    cur.execute("CREATE TABLE SemanalesIBEX35 (Ticker TEXT REFERENCES Datos(Ticker), Fecha TEXT, Price DOUBLE, Low DOUBLE, High DOUBLE)")
    for fila in tabla[1:]:
        cur.execute("INSERT INTO SemanalesIBEX35 VALUES (?,?,?,?,?)", (fila[0], fila[1].split()[0], fila[2], fila[3], fila[4]))   
    cur.close()
    conn.commit()

def crearBaseDatos():
    leerDatosGenerales()
    leerDatosIbex()
    leerDatosDAX30()
    leerDatosNasdaq100()
    leerDatosFtseMib()
    leerDatosSemanalesIBEX()

def consulta1():
    conn=sqlite3.connect("bolsa.sqlite3")
    cur=conn.cursor()

    query = """SELECT Nombre, Indice, Max - Min FROM
               (SELECT * from DAX30
                       UNION SELECT * FROM FtseMib
                       UNION SELECT * FROM Ibex
                       UNION SELECT * FROM Nasdaq100)
                       JOIN Datos USING(Ticker)
               WHERE Max - Min == (SELECT MAX(Max - Min) FROM
                                       (SELECT * from DAX30
                                           UNION SELECT * FROM FtseMib
                                           UNION SELECT * FROM Ibex
                                           UNION SELECT * FROM Nasdaq100));"""

    for row in cur.execute(query):
        print(row)

    cur.close()

def consulta2():
    conn=sqlite3.connect("bolsa.sqlite3")
    cur=conn.cursor()

    query = """SELECT Indice, Pais, SUM(Volumen) FROM
                Datos JOIN (SELECT * from DAX30
                    UNION SELECT * FROM FtseMib
                    UNION SELECT * FROM Ibex
                    UNION SELECT * FROM Nasdaq100) USING(Ticker)
                GROUP BY Indice;"""

    for row in cur.execute(query):
        print(row)

    cur.close()

def consulta3():
    pass

def consulta4():
    pass

def consulta5():
    pass

def consulta6():
    pass

def consulta7():
    pass

```
