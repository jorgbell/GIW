```python
import sqlite3
import csv

def leerDatosGenerales():
    csvarchivo = open('Tabla1.csv',encoding="iso-8859-1",errors='ignore')
    entrada = csv.reader(csvarchivo, delimiter=";")
    tabla = list(entrada)
    conn=sqlite3.connect("bolsa.sqlite3")
    cur=conn.cursor()
    cur.execute("DROP TABLE IF EXISTS Datos")
    cur.execute("CREATE TABLE Datos (Ticker TEXT UNIQUE, Nombre TEXT, Indice TEXT, Pais TEXT)")
    for fila in tabla[1:]:
        cur.execute("INSERT INTO Datos VALUES (?,?,?,?)", fila)   
    cur.close()
    conn.commit()

def leerDatosIbex():
    csvarchivo = open('Tabla2(IBEX35).csv',encoding="iso-8859-1",errors='ignore')
    entrada = csv.reader(csvarchivo, delimiter=";")
    tabla = list(entrada)
    conn=sqlite3.connect("bolsa.sqlite3")
    cur=conn.cursor()
    cur.execute("DROP TABLE IF EXISTS Ibex")
    cur.execute("CREATE TABLE Ibex (Ticker TEXT REFERENCES Datos(Ticker) UNIQUE, Precio DECIMAL, VarPorcentaje DECIMAL, VarAbsoluto DECIMAL, Max DECIMAL, Min DECIMAL, Volumen INTEGER)")
    for fila in tabla[1:]:
        cur.execute("INSERT INTO Ibex VALUES (?,?,?,?,?,?,?)", fila)   
    cur.close()
    conn.commit()


def leerDatosDAX30():
    csvarchivo = open('Tabla3(DAX 30).csv',encoding="iso-8859-1",errors='ignore')
    entrada = csv.reader(csvarchivo, delimiter=";")
    tabla = list(entrada)
    conn=sqlite3.connect("bolsa.sqlite3")
    cur=conn.cursor()
    cur.execute("DROP TABLE IF EXISTS DAX30")
    cur.execute("CREATE TABLE DAX30 (Ticker TEXT REFERENCES Datos(Ticker) UNIQUE, Precio DECIMAL, VarPorcentaje DECIMAL, VarAbsoluto DECIMAL, Max DECIMAL, Min DECIMAL, Volumen INTEGER)")
    for fila in tabla[1:]:
        cur.execute("INSERT INTO DAX30 VALUES (?,?,?,?,?,?,?)", fila)
    cur.close()
    conn.commit()
    
def leerDatosNasdaq100():
    csvarchivo = open('Tabla4(Nasdaq 100).csv',encoding="iso-8859-1",errors='ignore')
    entrada = csv.reader(csvarchivo, delimiter=";")
    tabla = list(entrada)
    conn=sqlite3.connect("bolsa.sqlite3")
    cur=conn.cursor()
    cur.execute("DROP TABLE IF EXISTS Nasdaq100")
    cur.execute("CREATE TABLE Nasdaq100 (Ticker TEXT REFERENCES Datos(Ticker) UNIQUE, Precio DECIMAL, VarPorcentaje DECIMAL, VarAbsoluto DECIMAL, Max DECIMAL, Min DECIMAL, Volumen INTEGER)")
    for fila in tabla[1:]:
        cur.execute("INSERT INTO Nasdaq100 VALUES (?,?,?,?,?,?,?)", fila)   
    cur.close()
    conn.commit()
    
def leerDatosFtseMib():
    csvarchivo = open('Tabla5(Ftse Mib).csv',encoding="iso-8859-1",errors='ignore')
    entrada = csv.reader(csvarchivo, delimiter=";")
    tabla = list(entrada)
    conn=sqlite3.connect("bolsa.sqlite3")
    cur=conn.cursor()
    cur.execute("DROP TABLE IF EXISTS FtseMib")
    cur.execute("CREATE TABLE FtseMib (Ticker TEXT REFERENCES Datos(Ticker) UNIQUE, Precio DECIMAL, VarPorcentaje DECIMAL, VarAbsoluto DECIMAL, Max DECIMAL, Min DECIMAL, Volumen INTEGER)")
    for fila in tabla[1:]:
        cur.execute("INSERT INTO FtseMib VALUES (?,?,?,?,?,?,?)", fila)   
    cur.close()
    conn.commit()
def leerDatosSemanalesIBEX():
    csvarchivo = open('Tabla6(SemanalesIBEX35).csv',encoding="iso-8859-1",errors='ignore')
    entrada = csv.reader(csvarchivo, delimiter=";")
    tabla = list(entrada)
    conn=sqlite3.connect("bolsa.sqlite3")
    cur=conn.cursor()
    cur.execute("DROP TABLE IF EXISTS SemanalesIBEX35")
    cur.execute("CREATE TABLE SemanalesIBEX35 (Ticker TEXT REFERENCES Datos(Ticker), Fecha TEXT, Price DECIMAL, Low DECIMAL, High DECIMAL)")
    for fila in tabla[1:]:
        cur.execute("INSERT INTO SemanalesIBEX35 VALUES (?,?,?,?,?)", (fila[0], fila[1].split()[0], fila[2].replace(',', '.'), fila[3].replace(',', '.'), fila[4].replace(',', '.')))   
    cur.close()
    conn.commit()

def crearBaseDatos():
    leerDatosGenerales()
    leerDatosIbex()
    leerDatosDAX30()
    leerDatosNasdaq100()
    leerDatosFtseMib()
    leerDatosSemanalesIBEX()

def ejecutarConsulta(query):
    conn=sqlite3.connect("bolsa.sqlite3")
    cur=conn.cursor()

    for row in cur.execute(query):
        print(row)

    cur.close()

def consulta1():
    query = """SELECT Nombre, Indice, MAX(Max - Min) FROM
                (SELECT * from DAX30
                    UNION SELECT * FROM FtseMib
                    UNION SELECT * FROM Ibex
                    UNION SELECT * FROM Nasdaq100)
                JOIN Datos USING(Ticker);"""

    ejecutarConsulta(query)

def consulta2():
    query = """SELECT Indice, Pais, SUM(Volumen) AS VolTotal FROM
                Datos JOIN (SELECT * from DAX30
                    UNION SELECT * FROM FtseMib
                    UNION SELECT * FROM Ibex
                    UNION SELECT * FROM Nasdaq100) USING(Ticker)
                GROUP BY Indice
                ORDER BY VolTotal DESC;"""

    ejecutarConsulta(query)

def consulta3():
    query = """SELECT Ticker, (Max - Min) AS MaxMinVariation, LowHighVariation FROM Ibex
                    JOIN (SELECT Ticker, MIN(High - Low) AS "LowHighVariation" FROM
                        SemanalesIBEX35
                        GROUP BY Ticker) USING(Ticker)
                WHERE MaxMinVariation < LowHighVariation;"""

    ejecutarConsulta(query)

def consulta4():
    query = """SELECT Ticker, AVG(Price) AS Media FROM
                SemanalesIBEX35
                GROUP BY Ticker
                ORDER BY Media"""

    ejecutarConsulta(query)

def consulta5():
    queryMaxPrecio = """SELECT Nombre, Indice, Pais, MAX(Precio) FROM
                        Datos JOIN (SELECT * from DAX30
                                    UNION SELECT * FROM FtseMib
                                    UNION SELECT * FROM Ibex
                                    UNION SELECT * FROM Nasdaq100) USING(Ticker);"""

    queryMaxVolumen = """SELECT Nombre, Indice, Pais, MAX(Volumen) FROM
                         Datos JOIN (SELECT * from DAX30
                                     UNION SELECT * FROM FtseMib
                                     UNION SELECT * FROM Ibex
                                     UNION SELECT * FROM Nasdaq100) USING(Ticker);"""

    print('Accion con mayor precio:')
    ejecutarConsulta(queryMaxPrecio)
    
    print('Accion con mayor volumen negociado:')
    ejecutarConsulta(queryMaxVolumen)
 
def consulta6():
    query = """SELECT t1.Nombre, t2.Nombre,  t1.Fecha, t1.Price - t2.Price AS Diferencia FROM
               (SemanalesIBEX35 JOIN Datos USING(Ticker)) t1
                   JOIN (SemanalesIBEX35 JOIN Datos USING(Ticker)) t2 ON (t1.Ticker != t2.Ticker AND t1.Fecha == t2.Fecha)
               WHERE Diferencia > 0
               ORDER BY t1.Fecha;"""
    
    ejecutarConsulta(query)

def consulta7():
    query = """SELECT Nombre, Fecha FROM
               SemanalesIBEX35
                   JOIN Datos USING(Ticker)
               WHERE Price - High == 0
                   AND Fecha == (SELECT fecha FROM
                                   (SELECT fecha, MAX(contador) FROM
                                       (SELECT *, COUNT(*) AS Contador FROM
                                       SemanalesIBEX35 t1
                                       WHERE Price - High == 0
                                       GROUP BY Fecha)));"""

    ejecutarConsulta(query)

```
