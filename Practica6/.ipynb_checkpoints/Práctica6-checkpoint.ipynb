### Práctica sobre APIs

Se va a explotar la API REST VACCOVID - coronavirus, vaccine and treatment tracker que se encuentra en:

https://rapidapi.com/vaccovidlive-vaccovidlive-default/api/vaccovid-coronavirus-vaccine-and-treatment-tracker/

Lo primero que se debe hacer es registrarse en el siguiente enlace:
https://rapidapi.com/auth/sign-up?referral=/hub

Una vez registrado y autenticado en la API REST se puede volver al endpoint donde se podrán ver los diferentes grupos de funcionalidad tales como:

* covid-19 APIs

* Vaccine and Treatments APIs

* News

En cada grupo se encuentran definidos varios conjuntos de servicios. Cuando se selecciona un servicio aparece la forma de invocar dicho servicio. En particular hay una pestaña en la pantalla denominada "Code Snippets" donde se puede encontrar el código para hacer uso del servicio en diferentes lenguajes de programación. Para ello , se busca en el desplegable. En este caso, se busca Python, y entre las opciones que ofrece se selecciona la opción "requests". Como resultado, se muestra el código en Python. Por ejemplo, si se quiere recuperar las estadísticas de covid-19 de los últimos 6 meses de Estados Unidos, bastaría pulsar sobre covid-19 APIs y allí sobre el servicio "get the covid-19 stats for the past six months of one country".En el marco que hay al lado del menú, se configura la api, y se rellena el apartado "Required Parameters" introduciendo el valor de id igual a "USA". En la pestaña "Code Snippets" se busca el lenguaje Python, y se selecciona la versión "requests". Entonces aparece el siguiente código:

```python
import requests

url = "https://vaccovid-coronavirus-vaccine-and-treatment-tracker.p.rapidapi.com/api/covid-ovid-data/sixmonth/USA"

headers = {
	"X-RapidAPI-Key": "4d136bbbafmshcef7c38206f927dp13110djsnb1ce888ae541",
	"X-RapidAPI-Host": "vaccovid-coronavirus-vaccine-and-treatment-tracker.p.rapidapi.com"
}

response = requests.request("GET", url, headers=headers)

print(response.text)
```

Al ejecutarse nos devuelve una cadena que representa un cadena json.

Observar que en la invocación de la API aparece una clave de acceso "x-rapidapi-key". Esta clave es exclusiva de cada usuario. Por cada grupo, basta que un de los componentes se registre en la aplicación. No hace falta que os registréis todos.

```python
import pprint
import requests
import json

def getResponse(url, key):
    
    headers = {
        "X-RapidAPI-Key": key,
        "X-RapidAPI-Host": "vaccovid-coronavirus-vaccine-and-treatment-tracker.p.rapidapi.com"
    }
    
    response = requests.request("GET", url, headers=headers)
    
    j = json.loads(response.text)
    return j
```

### OPCION 1

```python
def muestraDatosCountry(c):
    ##resultado
    print('\nInfection_Risk: ', c['Infection_Risk'])
    print('Case_Fatality_Rate: ', c['Case_Fatality_Rate'])
    print('Test_Percentage: ', c['Test_Percentage'])
    print('Recovery_Proporation: ', c['Recovery_Proporation'])
    print('TotalCases: ', c['TotalCases'])
    print('NewCases: ', c['NewCases'])
    print('TotalDeaths: ', c['TotalDeaths'])
    print('NewDeaths: ', c['NewDeaths'])
    print('TotalRecovered: ', c['TotalRecovered'])
    print('NewRecovered: ', c['NewRecovered'])
    print('ActiveCases: ', c['ActiveCases'])
    print('TotalTests: ', c['TotalTests'])
    print('Population: ', c['Population'])


def opcion1():
    otravez = True
    while otravez:
        print("\n¿Qué continente quieres?")
        url = "https://vaccovid-coronavirus-vaccine-and-treatment-tracker.p.rapidapi.com/api/npm-covid-data/"
        
        #pillamos el json
        rjson = getResponse(url, "6ee95f1d0fmsh0967d26374f8118p188173jsnc79ea1b2dc6a")
        
        continents = []
        i = 0
        for c in rjson:
            cont = c['Continent']
            if cont not in continents:
                continents.append(cont)
                print(i, cont)
                i+=1
        
        contElegido = int(input())
        
        print("\n¿Qué pais de ese continente quieres?")
        countries = []
        i = 0
        for c in rjson:
            cont = c['Continent']
            if cont == continents[contElegido]:
                country = c['Country']
                countries.append(country)
                print(i, country)
                i+=1
        
        countryElegida = int(input())
        
        for c in rjson:
            cont = c['Continent']
            country = c['Country']
            if cont == continents[contElegido] and country == countries[countryElegida]:
                muestraDatosCountry(c)
                break
        
        #siguiente consulta
        print('\n¿Quieres hacer otra consulta?')
        print('0: NO')
        print('1: SI')
        otravez = int(input())
        
    menuPrincipal()
```

### OPCION 2

```python
def opcion2():
    otravez = True
    while otravez:
        print("\n¿Qué información quieres?")
        print("0: Treatment")
        print("1: Vaccine")
        
        r = input()
        if(r == 0): 
            url = "https://vaccovid-coronavirus-vaccine-and-treatment-tracker.p.rapidapi.com/api/vaccines/get-all-treatment" 
        else:
            url = "https://vaccovid-coronavirus-vaccine-and-treatment-tracker.p.rapidapi.com/api/vaccines/get-all-vaccines"
        #pillamos el json
        rjson = getResponse(url, "1797033e79msh9ef00786d6b0d3dp1aed70jsnebc185f18a8a")
        #mostramos todas las categorias
        categories = []
        for item in rjson:
            c = item['trimedCategory']
            if c not in categories:
                categories.append(c)
            
        print("\nEscoge la categoria:")
        i = 0
        for c in categories:
            print(i,c)
            i+=1
        #elige la categoria
        catElegida = categories[int(input())]
        #filtramos el json segun la categoria
        rjson = list(filter((lambda item : item['trimedCategory'] == catElegida), rjson))
        
        #mostramos los laboratorios
        print("\nEscoge el laboratorio:")
        i = 0
        for item in rjson:
            print(i, item['developerResearcher'])
            i+=1
    
        labElegido = rjson[int(input())]
        
        #resultado
        print('\ndeveloperResearcher: ', labElegido['developerResearcher'])
        print('trimedName: ', labElegido['trimedName'])
        print('phase: ', labElegido['phase'])
        print('nextSteps: ', labElegido['nextSteps'])
        print('description: ', labElegido['description'])
        print('funder: ', labElegido['funder'])
        
        #siguiente consulta
        print('\n¿Quieres hacer otra consulta?')
        print('0: NO')
        print('1: SI')
        otravez = int(input())

    menuPrincipal()
```

### OPCION 3

```python
import webbrowser

def opcion3():
    otraVez = True
    
    while otraVez:    
        url = 'https://vaccovid-coronavirus-vaccine-and-treatment-tracker.p.rapidapi.com/api/news/get-coronavirus-news/0'
        rjson = getResponse(url, "1797033e79msh9ef00786d6b0d3dp1aed70jsnebc185f18a8a")['news']

        # Menu elegir fuente de informacion
        fuentesInformacion = []
        for noticia in rjson:
            if noticia['reference'] not in fuentesInformacion:
                fuentesInformacion.append(noticia['reference'])

        print('Elige la fuente de informacion que deseas: ')
        i = 0
        for f in fuentesInformacion:
            print(i, f)
            i+=1

        fuenteElegida = fuentesInformacion[int(input('\nIntroduce el numero asociado a la fuente de informacion: '))]

        rjson = list(filter((lambda noticia: noticia['reference'] == fuenteElegida), rjson))

        # Menu elegir noticia
        print('\nElige la noticia que deseas: ')
        i = 0
        for noticia in rjson:
            print(i, noticia['title'])
            i+=1

        noticiaElegida = list(rjson)[int(input('\nIntroduce el numero asociado a la noticia: '))]

        # Mostrar 'content'
        print('\n', noticiaElegida['content'])

        # Menu ver pagina de la noticia
        print('\nQuieres ver la pagina web de la noticia?')
        print('0: NO')
        print('1: SI')
        verPagina = int(input())

        if verPagina:
            link = noticiaElegida['link']
            webbrowser.open(link)

        #siguiente consulta
        print('\n¿Quieres hacer otra consulta?')
        print('0: NO')
        print('1: SI\n')
        otraVez = int(input())
    
    menuPrincipal()
```
