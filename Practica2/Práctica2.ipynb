A continuación, debeis calcular la distancia entre cada monumento y el lugar indicado. Esto se puede hacer usando la siguiente función que os propociono. 

```python
from math import radians, cos, sin, asin, sqrt
def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    # Radius of earth in kilometers is 6371
    km = 6371* c
    return km
#Ejemplo de distancia entre la facultad y el Ministerio de Agricultura
haversine (-3.6906142029315534,40.408789366623914,  -3.7371139,40.4520948)
```

```python
def consulta1(accidentes, nameColIndex):
    dicNumAccDisTipo = dict()
    for i in range(1, len(accidentes)):
        nombreDistrito = accidentes[i][nameColIndex["distrito"]]
        tipoAccidente = accidentes[i][nameColIndex["tipo_accidente"]]
        
        if dicNumAccDisTipo.get((nombreDistrito, tipoAccidente)) != None:
            dicNumAccDisTipo[(nombreDistrito, tipoAccidente)] += 1
        else:
            dicNumAccDisTipo[(nombreDistrito, tipoAccidente)] = 1
    
    return dicNumAccDisTipo
```

```python
import csv
import pprint as p
csvarchivo = open('AccidentesBicicletas_2021.csv',encoding="utf8",errors='ignore')
entrada = csv.reader(csvarchivo, delimiter=";")
    
accidentes = list(entrada)

nameColIndex = dict()
for i in range(len(accidentes[0])):
    nameColIndex[accidentes[0][i]] = i
    
p.pprint(consulta1(accidentes, nameColIndex))
```

```python
def consulta2A(accidentes, nameColIndex, fecha):
    dicNumAccDisTipo = dict()
    for i in range(1, len(accidentes)):
        #comprueba si es la fecha indicada
        #asumimos que no tiene por que estar ordenado el csv, por lo que busca la fecha en todo el documento
        if accidentes[i][nameColIndex["fecha"]] != fecha:
            continue
           
        nombreDistrito = accidentes[i][nameColIndex["distrito"]]
        tipoLesividad = accidentes[i][nameColIndex["tipo_lesividad"]]
        
        if dicNumAccDisTipo.get(tipoLesividad) != None:
            if dicNumAccDisTipo[tipoLesividad].get(nombreDistrito) != None:
                dicNumAccDisTipo[tipoLesividad][nombreDistrito] += 1
            else:
                dicNumAccDisTipo[tipoLesividad][nombreDistrito] = 1
        else: #si no existe el tipo de lesividad aun
            dicNumAccDisTipo[tipoLesividad] = {nombreDistrito: 1}
            
    results = []
    #funcion que devuelve el distrito con mayor numero de accidentes segun el tipo de lesividad
    #si es igual, devuelve el primero de la lista
    for clave,valor in dicNumAccDisTipo.items():
        maxDistrito = max(valor.items(), key = (lambda item : item[1])) #funcion comparadora para comparar items del diccionario
        results.append((clave,maxDistrito))
    
    return results
```

```python
#Consulta 2A

dia = input("Escribe el número del día con 2 dígitos: ")
mes = input("Escribe el número del mes con 2 dígitos: ")

fecha = "/".join([dia, mes, "2021"])
print(fecha)
consulta2A(accidentes, nameColIndex, fecha)
```
